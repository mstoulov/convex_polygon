В файле 'square.txt' записаны количества многоугольников для n от 2 до 1000 и отношения соседних количеств.

Оставим только количества, построим график и зная, что асимптотика имеет вид a*b^(n^(2/3)), попытаемся найти a и b

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('square.txt', sep=' ')
```

```python
df = df.rename(columns={'1' : 'sz', '0' : 'cnt'})
df = df.reset_index()
df = df[df.index < 999]
df = df.set_index('sz')
df = df.drop(columns=['index'])
df
```

```python
plt.figure(figsize=(20, 10))
plt.plot(np.log(df.cnt), label='data')
plt.plot(np.power(np.arange(1, 1000), (2 / 3)) * 6.76 - 14, label='n^2/3')
#plt.xlim((100, 1010))
plt.legend()
plt.show()
```

таким образом, если поверить в то, что реальные значения сходятся к асимптотике довольно быстро 
(в чем я совершенно не уверен), можно сказать, что асимптотика имеет вид 
(e^-14) * (e^6.76)^(n^(2/3)) = 8e-7 * 863^(n^(2/3))
